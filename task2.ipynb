```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img_path = "images\cube.jpg"
```

## setting image to grayscale and appling gaussian blur

```python
img = cv2.imread(img_path)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gauss = cv2.GaussianBlur(gray,(5,5),50)
plt.imshow(gauss,cmap="gray")
plt.axis("off")
plt.show()
```

## edge detection using Laplacian

```python
laplacian = cv2.Laplacian(gray, cv2.CV_64F)
laplacian = np.uint8(np.absolute(laplacian))
cv2.imwrite("edge_images\laplacian_cube.jpg",laplacian)
plt.imshow(laplacian,cmap="gray")
plt.axis("off")
plt.show()
```

## edge detection using Sobel

```python
sobel_x = cv2.Sobel(gauss,-1,1,0)
sobel_y = cv2.Sobel(gauss,-1,0,1)
sobel = cv2.bitwise_or(sobel_x,sobel_y)
cv2.imwrite("edge_images\sobel_cube.jpg",sobel)
plt.imshow(sobel,cmap="gray")
plt.axis("off")
plt.show()
```

## edge detection using Canny

```python
canny = cv2.Canny(img,100,200)
cv2.imwrite("edge_images\canny_cube.jpg",canny)
plt.imshow(canny,cmap="gray")
plt.axis("off")
plt.show()
```

## Other examples

```python
cat_path = "images\\normal_cat.jpg"
cat = cv2.imread(cat_path)

cat = cv2.cvtColor(cat,cv2.COLOR_BGR2RGB)
plt.subplot(221)
plt.imshow(cat)
plt.title("original")
plt.axis("off")


gray = cv2.cvtColor(cat,cv2.COLOR_RGB2GRAY)
gauss = cv2.GaussianBlur(gray,(7,7),75)

laplacian = cv2.Laplacian(gray, cv2.CV_64F)
laplacian = np.uint8(np.absolute(laplacian))
cv2.imwrite("edge_images\laplacian_cat.jpg",laplacian)

plt.subplot(222)
plt.imshow(laplacian,cmap="gray")
plt.title("gray")
plt.axis("off")


sobel_x = cv2.Sobel(gauss,-1,1,0)
sobel_y = cv2.Sobel(gauss,-1,0,1)
sobel = cv2.bitwise_or(sobel_x,sobel_y)
cv2.imwrite("edge_images\sobel_cat.jpg",sobel)

plt.subplot(223)
plt.imshow(sobel,cmap="gray")
plt.title("sobel")
plt.axis("off")


canny = cv2.Canny(gauss,50,100)
cv2.imwrite("edge_images\canny_cat.jpg",canny)

plt.subplot(224)
plt.imshow(canny,cmap="gray")
plt.title("canny")
plt.axis("off")

plt.show()
```
